```python
import cv2
import numpy as np
import os
from ultralytics import YOLO

# تحميل النماذج
pose_model = YOLO('yolov8n-pose.pt')         # keypoints
ball_model = YOLO('yolov8n.pt')              # ball detection

# ----------------------------------------
# 1. حساب ارتفاع القفز وعدد الإطارات في الهواء
# ----------------------------------------
def evaluate_jump_height_and_air_time(video_path, frame_skip=2, height_thresh=150):
    cap = cv2.VideoCapture(video_path)
    frame_idx = 0
    jump_frames = []
    jump_heights = []

    print("🕵️‍♀️ تحليل القفز بناءً على ارتفاع القدم ووضعية الجسم...\n")

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        if frame_idx % frame_skip != 0:
            frame_idx += 1
            continue

        # كشف النقاط المفتاحية
        pose_results = pose_model(frame)[0]
        if pose_results.keypoints is not None and pose_results.keypoints.xy.shape[0] > 0:
            keypoints = pose_results.keypoints.xy.cpu().numpy()[0]

            if keypoints.shape[0] >= 17:
                hip = keypoints[11]  # الورك الأيسر
                ankle = keypoints[15]  # القدم اليسرى

                # حساب الارتفاع بين الورك والقدم
                height_diff = ankle[1] - hip[1]
                jump_heights.append(height_diff)

                print(f"📽️ إطار {frame_idx} - ارتفاع القفزة: {round(height_diff, 1)}px")

                # تقدير أن اللاعب "في الهواء"
                if height_diff > height_thresh:
                    jump_frames.append(frame_idx)
                    print("✅ اللاعب في الهواء!")

        frame_idx += 1

    cap.release()

    total_jump_frames = len(jump_frames)
    max_jump = round(max(jump_heights), 1) if jump_heights else 0

    print("\n📊 النتائج:")
    print(f"🔺 أعلى قفزة: {max_jump}px")
    print(f"🕐 عدد الإطارات في الهواء: {total_jump_frames}")
    print(f"📈 متوسط الارتفاع: {round(np.mean(jump_heights), 1)}px" if jump_heights else "لا يوجد بيانات.")

    return {
        "max_jump_height": max_jump,
        "air_frames": total_jump_frames,
        "jump_frame_indices": jump_frames
    }

# ----------------------------------------
# 2. رسم النقاط المفتاحية وصندوق الكرة
# ----------------------------------------
def draw_keypoints_and_ball(frame, pose_results, ball_results):
    if pose_results.keypoints is not None and pose_results.keypoints.xy.shape[0] > 0:
        keypoints = pose_results.keypoints.xy.cpu().numpy()[0]
        for point in keypoints:
            x, y = int(point[0]), int(point[1])
            cv2.circle(frame, (x, y), 4, (0, 255, 0), -1)  # نقطة خضراء

    for box, cls in zip(ball_results.boxes.xyxy, ball_results.boxes.cls):
        if int(cls) in [32, 37]:  # sports ball or frisbee
            x1, y1, x2, y2 = map(int, box)
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 0, 255), 2)
            cv2.putText(frame, "Ball", (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)

    return frame

# ----------------------------------------
# 3. حفظ الإطارات التي كان فيها اللاعب في الهواء
# ----------------------------------------
def save_jump_frames(video_path, frame_indices, output_dir="jump_frames"):
    os.makedirs(output_dir, exist_ok=True)

    cap = cv2.VideoCapture(video_path)
    frame_idx = 0
    saved = 0

    print(f"\n💾 جاري حفظ {len(frame_indices)} إطارًا في المجلد: '{output_dir}'")

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        if frame_idx in frame_indices:
            pose_results = pose_model(frame)[0]
            ball_results = ball_model(frame)[0]

            frame_annotated = draw_keypoints_and_ball(frame, pose_results, ball_results)
            filename = f"{output_dir}/jump_frame_{frame_idx}.jpg"
            cv2.imwrite(filename, frame_annotated)
            print(f"✅ تم حفظ: {filename}")
            saved += 1

        frame_idx += 1
        if saved >= len(frame_indices):
            break

    cap.release()
    print("🎉 تم حفظ كل الإطارات المطلوبة!")

# ----------------------------------------
# 🧪 الاستخدام:
# ----------------------------------------
video_path = "WhatsApp Video 2025-04-05 at 21.58.18.mp4"

results = evaluate_jump_height_and_air_time(video_path)
save_jump_frames(video_path, results["jump_frame_indices"])
```

**1- Jumping with ball**

```python
import cv2
import numpy as np
import os
from ultralytics import YOLO

# Load models
pose_model = YOLO("yolov8n-pose.pt")
ball_model = YOLO("yolov8n.pt")

def evaluate_jump_height_and_air_time(video_path, frame_skip=2, height_thresh=150, air_thresh=30):
    cap = cv2.VideoCapture(video_path)
    frame_idx = 0
    jump_frames = []
    jump_heights = []

    os.makedirs("output_frames", exist_ok=True)
    print("🕵️‍♀️ تحليل القفز والكشف عن الكرة...\n")

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        if frame_idx % frame_skip != 0:
            frame_idx += 1
            continue

        # Run pose model
        pose_results = pose_model(frame)[0]
        keypoints_detected = False
        jump_in_air = False
        height_diff = 0

        if pose_results.keypoints is not None and pose_results.keypoints.xy.shape[0] > 0:
            keypoints = pose_results.keypoints.xy.cpu().numpy()[0]

            if keypoints.shape[0] >= 17:
                hip = keypoints[11]
                ankle = keypoints[15]
                height_diff = ankle[1] - hip[1]
                jump_heights.append(height_diff)

                if height_diff > height_thresh:
                    jump_frames.append(frame_idx)
                    jump_in_air = True

                keypoints_detected = True
                for x, y in keypoints:
                    cv2.circle(frame, (int(x), int(y)), 5, (0, 255, 0), -1)

        # Run ball detection and filter properly
        ball_results = ball_model(frame)[0]
        for box in ball_results.boxes:
            conf = box.conf[0].item()
            cls_id = int(box.cls[0])
            label = ball_model.names[cls_id]

            if label.lower() == 'ball' and conf > 0.5:
                x1, y1, x2, y2 = map(int, box.xyxy[0])
                cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 0, 255), 2)
                cv2.putText(frame, "Ball", (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 0, 255), 2)

        # Annotate frame
        cv2.putText(frame, f"Frame: {frame_idx}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2)
        if keypoints_detected:
            cv2.putText(frame, f"Jump Height: {height_diff:.2f} px", (10, 60), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2)
        if jump_in_air:
            cv2.putText(frame, "In Air!", (10, 90), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 255), 2)

        # Save the annotated frame
        cv2.imwrite(f"output_frames/frame_{frame_idx}.jpg", frame)

        frame_idx += 1

    cap.release()

    # Final evaluation
    total_jump_frames = len(jump_frames)
    max_jump = round(max(jump_heights), 1) if jump_heights else 0
    avg_jump = round(np.mean(jump_heights), 1) if jump_heights else 0

    # Score calculation
    score = 0
    if max_jump > height_thresh:
        score += 2
    if total_jump_frames > air_thresh:
        score += 2
    if avg_jump > (height_thresh * 0.8):
        score += 1

    print("\n📊 النتائج:")
    print(f"🔺 أعلى قفزة: {max_jump}px")
    print(f"📈 متوسط الارتفاع: {avg_jump}px")
    print(f"🕐 عدد الإطارات في الهواء: {total_jump_frames}")
    print(f"⭐ تقييم الأداء: {score} / 5")

    return {
        "max_jump_height": max_jump,
        "avg_jump_height": avg_jump,
        "air_frames": total_jump_frames,
        "jump_frame_indices": jump_frames,
        "score": score
    }

# Example usage
results = evaluate_jump_height_and_air_time("WhatsApp Video 2025-04-05 at 21.58.18.mp4")
```

**2-Running with Ball**

```python
import cv2
import numpy as np
import os
from ultralytics import YOLO

# Load pose estimation model
pose_model = YOLO("yolov8n-pose.pt")

def evaluate_running_pose_only(video_path, frame_skip=2, min_distance=100):
    cap = cv2.VideoCapture(video_path)
    frame_idx = 0
    mid_hip_positions = []
    left_ankle_y = []
    right_ankle_y = []
    step_changes = 0

    os.makedirs("pose_run_frames", exist_ok=True)

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        if frame_idx % frame_skip != 0:
            frame_idx += 1
            continue

        # Pose estimation
        pose_results = pose_model(frame)[0]
        if pose_results.keypoints is not None and pose_results.keypoints.xy.shape[0] > 0:
            keypoints = pose_results.keypoints.xy.cpu().numpy()[0]

            if keypoints.shape[0] >= 17:
                # Mid-hip (average of left and right hip)
                mid_hip = (keypoints[11] + keypoints[12]) / 2
                mid_hip_positions.append(mid_hip)

                # Track ankle height to estimate steps
                left_ankle_y.append(keypoints[15][1])
                right_ankle_y.append(keypoints[16][1])

                # Draw points
                cv2.circle(frame, (int(mid_hip[0]), int(mid_hip[1])), 5, (0, 255, 0), -1)
                cv2.putText(frame, f"Frame: {frame_idx}", (10, 25),
                            cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2)

        cv2.imwrite(f"pose_run_frames/frame_{frame_idx}.jpg", frame)
        frame_idx += 1

    cap.release()

    # Total movement distance (sum of mid-hip displacement)
    total_distance = 0
    for i in range(1, len(mid_hip_positions)):
        dx = mid_hip_positions[i][0] - mid_hip_positions[i - 1][0]
        dy = mid_hip_positions[i][1] - mid_hip_positions[i - 1][1]
        total_distance += np.sqrt(dx**2 + dy**2)

    # Estimate step count based on changes in ankle Y movement
    step_changes = 0
    for i in range(1, len(left_ankle_y)):
        if abs(left_ankle_y[i] - left_ankle_y[i - 1]) > 10 or abs(right_ankle_y[i] - right_ankle_y[i - 1]) > 10:
            step_changes += 1

    # Score based on distance and step activity
    score = 0
    if total_distance > min_distance:
        score += 3
    elif total_distance > (min_distance / 2):
        score += 2
    elif total_distance > (min_distance / 4):
        score += 1

    if step_changes > 5:
        score += 2
    elif step_changes > 2:
        score += 1

    score = min(score, 5)

    print("\n📊 نتائج الجري بناءً على الحركة:")
    print(f"📏 المسافة المقطوعة: {round(total_distance, 2)} px")
    print(f"🦶 عدد تغييرات حركة القدم (تقدير للخطوات): {step_changes}")
    print(f"✅ التقييم العام للجري (بدون كرة): {score}/5")

    return score

# Run it
evaluate_running_pose_only("running with ball.mp4")
```

# 3-Passing

```python
# ----- Skill 3: Passing -----
def evaluate_passing(video_path, frame_skip=2, ball_distance_thresh=100):
    cap = cv2.VideoCapture(video_path)
    frame_idx = 0
    pass_attempts = 0
    successful_passes = 0
    previous_ball_pos = None
    previous_mid_ankle = None

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        if frame_idx % frame_skip != 0:
            frame_idx += 1
            continue

        ball_results = ball_model(frame)[0]
        ball_pos = None
        for box, cls in zip(ball_results.boxes.xyxy, ball_results.boxes.cls):
            if int(cls) == 32:
                x1, y1, x2, y2 = map(int, box)
                ball_pos = ((x1 + x2) // 2, (y1 + y2) // 2)
                break

        pose_results = pose_model(frame)[0]
        if pose_results.keypoints is not None and pose_results.keypoints.xy.shape[0] > 0:
            keypoints = pose_results.keypoints.xy.cpu().numpy()[0]
            if keypoints.shape[0] >= 17:
                left_ankle = keypoints[15]
                right_ankle = keypoints[16]
                mid_ankle = (left_ankle + right_ankle) / 2

                if previous_ball_pos and ball_pos:
                    ball_move = np.linalg.norm(np.array(ball_pos) - np.array(previous_ball_pos))
                    ankle_to_ball = np.linalg.norm(np.array(ball_pos) - mid_ankle)
                    if ball_move > 20 and ankle_to_ball > ball_distance_thresh:
                        pass_attempts += 1
                        if ball_move < 150:
                            successful_passes += 1

                previous_ball_pos = ball_pos
                previous_mid_ankle = mid_ankle

        frame_idx += 1

    cap.release()
    pass_score = min(5, successful_passes)
    return pass_score

# ----- Run Passing Evaluation -----
def run_passing_evaluation_path(video_path):
    print("⚽️ جاري تقييم التمرير...")
    score = evaluate_passing(video_path)
    print(f"✅ درجة التمرير: {score}/5")
    return score
run_passing_evaluation_path("/content/9502550-sd_506_960_24fps.mp4")
```

```python
# ----- Skill 5: Zigzag Dribbling -----
def evaluate_zigzag_dribbling(video_path, frame_skip=2, direction_change_thresh=20):
    cap = cv2.VideoCapture(video_path)
    frame_idx = 0
    zigzag_changes = 0
    previous_mid_ankle = None
    last_direction = None

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        if frame_idx % frame_skip != 0:
            frame_idx += 1
            continue

        pose_results = pose_model(frame)[0]
        if pose_results.keypoints is not None and pose_results.keypoints.xy.shape[0] > 0:
            keypoints = pose_results.keypoints.xy.cpu().numpy()[0]
            if keypoints.shape[0] >= 17:
                left_ankle = keypoints[15]
                right_ankle = keypoints[16]
                mid_ankle = (left_ankle + right_ankle) / 2

                if previous_mid_ankle is not None:
                    dx = mid_ankle[0] - previous_mid_ankle[0]
                    if last_direction is not None and np.sign(dx) != np.sign(last_direction) and abs(dx) > direction_change_thresh:
                        zigzag_changes += 1
                    last_direction = dx

                previous_mid_ankle = mid_ankle

        frame_idx += 1

    cap.release()
    zigzag_score = min(5, zigzag_changes // 2)
    return zigzag_score

# ----- Run Zigzag Dribbling Evaluation -----
def run_zigzag_evaluation_path(video_path):
    print("🌀 جاري تقييم المراوغة الزجزاجية...")
    score = evaluate_zigzag_dribbling(video_path)
    print(f"✅ درجة المراوغة الزجزاجية: {score}/5")
    return score
```

```python
# --- Running Skill Evaluation (Ages 5-8) ---
import cv2
import numpy as np
import matplotlib.pyplot as plt
from ultralytics import YOLO

pose_model = YOLO('yolov8n-pose.pt')

def evaluate_running_skill(video_path):
    cap = cv2.VideoCapture(video_path)
    previous_position = None
    total_distance = 0
    frame_count = 0
    good_posture_frames = 0
    posture_threshold = 10

    speeds = []

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        results = pose_model(frame)
        frame_count += 1

        for result in results:
            if result.keypoints is not None and result.keypoints.xy.shape[1] >= 17:
                keypoints = result.keypoints.xy.cpu().numpy()[0]
                ankle = keypoints[15]
                hip = keypoints[11]

                if previous_position is not None:
                    dist = np.linalg.norm(ankle - previous_position)
                    total_distance += dist
                    speeds.append(dist)
                previous_position = ankle

                # posture angle (hip to ankle line vs vertical)
                delta_x = abs(ankle[0] - hip[0])
                delta_y = abs(ankle[1] - hip[1])
                angle = np.degrees(np.arctan2(delta_x, delta_y))

                if angle < posture_threshold:
                    good_posture_frames += 1

    cap.release()

    # Compute scores
    avg_speed = np.mean(speeds)
    posture_ratio = good_posture_frames / frame_count if frame_count > 0 else 0

    # Score out of 5
    speed_score = min(5, avg_speed * 10)
    posture_score = min(5, posture_ratio * 5)
    total_score = round((speed_score + posture_score) / 2, 2)

    print(f"\n🏃‍♂️ Running Speed Score: {speed_score:.2f}/5")
    print(f"🧍‍♂️ Posture Score: {posture_score:.2f}/5")
    print(f"⭐ Final Running Skill Score: {total_score}/5")

    # Visualization
    labels = ['Speed', 'Posture']
    values = [speed_score, posture_score]

    plt.figure(figsize=(6, 4))
    plt.bar(labels, values, color=['#4caf50', '#2196f3'])
    plt.ylim(0, 5)
    plt.title('Running Skill Evaluation (Ages 5-8)')
    plt.ylabel('Score (/5)')
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()

    return total_score
evaluate_running_skill("/content/running.mp4")
```

**2-Ball Feeling**

```python
# --- Ball Feeling Skill Evaluation (Ages 5-8) ---
def evaluate_ball_feeling_skill(video_path):
    cap = cv2.VideoCapture(video_path)
    contact_count = 0
    frame_count = 0

    prev_contact = False
    contact_threshold = 50  # distance threshold (pixels)

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        frame_count += 1
        pose_results = pose_model(frame)
        ball_results = ball_model(frame)

        foot = None
        ball = None

        for result in pose_results:
            if result.keypoints is not None and result.keypoints.xy.shape[1] >= 17:
                keypoints = result.keypoints.xy.cpu().numpy()[0]
                foot = keypoints[16]  # right foot (ankle)

        for r in ball_results:
            for box in r.boxes:
                if int(box.cls[0]) == 32:  # COCO class ID for sports ball
                    ball = box.xywh[0][:2].cpu().numpy()

        # Check if both detected
        if foot is not None and ball is not None:
            dist = np.linalg.norm(foot - ball)
            if dist < contact_threshold:
                if not prev_contact:
                    contact_count += 1
                    prev_contact = True
            else:
                prev_contact = False

    cap.release()
    # Score calculation
    if contact_count >= 10:
        score = 5
    elif contact_count >= 7:
        score = 4
    elif contact_count >= 5:
        score = 3
    elif contact_count >= 3:
        score = 2
    elif contact_count >= 1:
        score = 1
    else:
        score = 0

    print(f"⚽ Total Ball Contacts Detected: {contact_count}")
    print(f"⭐ Ball Feeling Skill Score: {score}/5")

    # Visualization
    labels = ['Ball Touches']
    values = [score]

    plt.figure(figsize=(5, 4))
    plt.bar(labels, values, color='orange')
    plt.ylim(0, 5)
    plt.title('Ball Feeling Evaluation')
    plt.ylabel('Score (/5)')
    plt.tight_layout()
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.show()

    return score
evaluate_ball_feeling_skill("/content/feel the ball.mp4")
```

```python
import cv2
import mediapipe as mp
import matplotlib.pyplot as plt

# --- Focus & Listening Skill Evaluation (Ages 5-8) ---
def evaluate_focus_listening_skill(video_path):
    mp_face_detection = mp.solutions.face_detection
    face_detection = mp_face_detection.FaceDetection(model_selection=0, min_detection_confidence=0.5)

    cap = cv2.VideoCapture(video_path)
    total_frames = 0
    attentive_frames = 0
    distracted_frames = 0

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        total_frames += 1
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        results = face_detection.process(frame_rgb)

        if results.detections:
            for detection in results.detections:
                # Get location data
                location = detection.location_data
                if location.relative_bounding_box.xmin < 0.4 or location.relative_bounding_box.xmin > 0.6:
                    distracted_frames += 1
                else:
                    attentive_frames += 1
        else:
            distracted_frames += 1

    cap.release()

    # Avoid divide by zero
    focus_ratio = attentive_frames / total_frames if total_frames > 0 else 0

    # Score logic
    if focus_ratio >= 0.8:
        score = 5
    elif focus_ratio >= 0.6:
        score = 4
    elif focus_ratio >= 0.4:
        score = 3
    elif focus_ratio >= 0.2:
        score = 2
    elif focus_ratio >= 0.05:
        score = 1
    else:
        score = 0

    print(f"📊 Attentive: {attentive_frames}, Distracted: {distracted_frames}")
    print(f"⭐ Focus & Listening Score: {score}/5")

    # Visualization
    plt.bar(['Focus'], [score], color='purple')
    plt.ylim(0, 5)
    plt.title('👂 Focus & Listening Skill Evaluation')
    plt.ylabel('Score (/5)')
    plt.tight_layout()
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.show()

    return score
evaluate_focus_listening_skill("/content/listening .mp4")
```

# *نجم لا يغيب - The ever-presented Stars

```python
# ✅ 1. Import Libraries
from ultralytics import YOLO
import numpy as np
import cv2
import pandas as pd

# ✅ 2. Load YOLOv8 Pose Model
model = YOLO('yolov8n-pose.pt')

# ✅ 3. Load Video File
video_path = "/content/الاصابة.mp4"

# ✅ 4. Extract Keypoints from Video
results = model(video_path, stream=True)
keypoints_sequence = []

for result in results:
    if result.keypoints is not None and result.keypoints.xy.shape[0] > 0:
        kps = result.keypoints.xy[0].cpu().numpy()
        keypoints_sequence.append(kps.tolist())

print(f"\n✅ Extracted keypoints from {len(keypoints_sequence)} frames 🎥")

# ✅ 5. Helper Functions
def calculate_angle(a, b, c):
    a, b, c = np.array(a), np.array(b), np.array(c)
    ba = a - b
    bc = c - b
    cosine_angle = np.dot(ba, bc) / (np.linalg.norm(ba) * np.linalg.norm(bc) + 1e-6)
    angle = np.arccos(np.clip(cosine_angle, -1.0, 1.0))
    return np.degrees(angle)

def calculate_distance(p1, p2):
    return np.linalg.norm(np.array(p1) - np.array(p2))

# ✅ 6. Biomechanics Feature Extraction
def get_knee_angle(keypoints):
    left_hip, left_knee, left_ankle = keypoints[11][:2], keypoints[13][:2], keypoints[15][:2]
    right_hip, right_knee, right_ankle = keypoints[12][:2], keypoints[14][:2], keypoints[16][:2]
    left_angle = calculate_angle(left_hip, left_knee, left_ankle)
    right_angle = calculate_angle(right_hip, right_knee, right_ankle)
    return left_angle, right_angle

def hip_flexion_angle(keypoints, side='right'):
    shoulder = keypoints[6][:2] if side == 'right' else keypoints[5][:2]
    hip = keypoints[12][:2] if side == 'right' else keypoints[11][:2]
    knee = keypoints[14][:2] if side == 'right' else keypoints[13][:2]
    return calculate_angle(shoulder, hip, knee)

def trunk_lean_angle(keypoints):
    left_shoulder, right_shoulder = np.array(keypoints[5][:2]), np.array(keypoints[6][:2])
    left_hip, right_hip = np.array(keypoints[11][:2]), np.array(keypoints[12][:2])
    shoulder_mid = (left_shoulder + right_shoulder) / 2
    hip_mid = (left_hip + right_hip) / 2
    vertical_ref = np.array([hip_mid[0], hip_mid[1] - 1])
    return calculate_angle(shoulder_mid, hip_mid, vertical_ref)

def thorax_rotation(keypoints):
    return keypoints[6][0] - keypoints[5][0]

def pelvic_tilt_asymmetry(keypoints):
    return keypoints[11][1] - keypoints[12][1]

def contact_angle(keypoints):
    ankle, knee = keypoints[15][:2], keypoints[13][:2]
    return calculate_angle(knee, ankle, [ankle[0], ankle[1] - 1])

def compute_velocity(positions, fps=30):
    return [(np.array(positions[i]) - np.array(positions[i-1])) * fps for i in range(1, len(positions))]

def compute_acceleration(velocities, fps=30):
    return [(np.array(velocities[i]) - np.array(velocities[i-1])) * fps for i in range(1, len(velocities))]

def step_lengths(positions, threshold=20):
    return [calculate_distance(positions[i], positions[i-1]) for i in range(1, len(positions)) if calculate_distance(positions[i], positions[i-1]) > threshold]

def step_frequency(step_count, total_time_seconds):
    return step_count / total_time_seconds

def asymmetry_score(left_value, right_value):
    return abs(left_value - right_value) / max(left_value, right_value + 1e-6) * 100

# ✅ 7. Feature Computation
hip_flexions, trunk_leans, thorax_rotations, pelvic_tilts = [], [], [], []
left_knees, right_knees, asymmetries, contact_angles = [], [], [], []
left_ankle_positions = [frame[15][:2] for frame in keypoints_sequence]

for kps in keypoints_sequence:
    l_knee, r_knee = get_knee_angle(kps)
    left_knees.append(l_knee)
    right_knees.append(r_knee)
    asymmetries.append(asymmetry_score(l_knee, r_knee))
    contact_angles.append(contact_angle(kps))
    hip_flexions.append(hip_flexion_angle(kps))
    trunk_leans.append(trunk_lean_angle(kps))
    thorax_rotations.append(thorax_rotation(kps))
    pelvic_tilts.append(pelvic_tilt_asymmetry(kps))

velocities = compute_velocity(left_ankle_positions)
accelerations = compute_acceleration(velocities)
acc_magnitudes = [np.linalg.norm(a) for a in accelerations]
steps = step_lengths(left_ankle_positions)
total_time = len(keypoints_sequence) / 30
cadence = step_frequency(len(steps), total_time)

# ✅ 8. Injury Risk Evaluation
def injury_risk_evaluation(knee_angle, asymmetry, contact_angle, max_acceleration, step_frequency):
    risk_score, notes = 0, []
    if knee_angle < 110 or knee_angle > 145:
        risk_score += 1; notes.append("⚠️ Knee angle out of safe range.")
    else:
        notes.append("✅ Knee angle within safe range.")
    if asymmetry > 15:
        risk_score += 2; notes.append("❌ High asymmetry — potential compensation risk.")
    elif asymmetry > 10:
        risk_score += 1; notes.append("⚠️ Moderate asymmetry.")
    else:
        notes.append("✅ Symmetry is good.")
    if contact_angle < 70 or contact_angle > 110:
        risk_score += 1; notes.append("⚠️ Risky landing posture.")
    else:
        notes.append("✅ Acceptable contact angle.")
    if max_acceleration > 500000:
        risk_score += 2; notes.append("❌ Very high acceleration — ligament risk.")
    elif max_acceleration > 250000:
        risk_score += 1; notes.append("⚠️ Elevated acceleration.")
    else:
        notes.append("✅ Acceleration within normal range.")
    if step_frequency < 1.5 or step_frequency > 3:
        risk_score += 1; notes.append("⚠️ Step frequency irregular.")
    else:
        notes.append("✅ Step frequency is good.")
    level = "🟢 Low" if risk_score <= 1 else "🟠 Medium" if risk_score <= 3 else "🔴 High"
    return {"level": level, "score": risk_score, "notes": notes}

def suspected_injury_type(knee_angle, asymmetry, contact_angle, max_acceleration, step_frequency, trunk_lean, hip_flexion):
    tags = []
    if hip_flexion > 35 or trunk_lean > 15:
        tags.append("⚠️ Hamstring overload (hip/trunk issue)")
    if knee_angle < 110 or knee_angle > 145:
        tags.append("⚠️ Knee joint stress (ACL risk)")
    if asymmetry > 15:
        tags.append("⚠️ Compensation injury (muscle imbalance)")
    if contact_angle < 70 or contact_angle > 110:
        tags.append("⚠️ Landing instability (ankle/knee)")
    if step_frequency < 1.5 or step_frequency > 3:
        tags.append("⚠️ Cadence issue (fatigue/coordination)")
    if max_acceleration > 500000:
        tags.append("⚠️ Acceleration overload (soft tissue risk)")
    return tags if tags else ["✅ No specific injury pattern"]

# ✅ 9. Final Summary
summary = {
    "Right Knee Angle Avg": round(np.mean(right_knees), 2),
    "Left Knee Angle Avg": round(np.mean(left_knees), 2),
    "Asymmetry Avg (%)": round(np.mean(asymmetries), 2),
    "Contact Angle Avg": round(np.mean(contact_angles), 2),
    "Max Acceleration": round(np.max(acc_magnitudes), 2),
    "Steps Count": len(steps),
    "Step Frequency": round(cadence, 2),
    "Hip Flexion Avg": round(np.mean(hip_flexions), 2),
    "Trunk Lean Avg": round(np.mean(trunk_leans), 2),
    "Pelvic Tilt Avg": round(np.mean(pelvic_tilts), 2),
    "Thorax Rotation Avg": round(np.mean(thorax_rotations), 2)
}

risk = injury_risk_evaluation(
    knee_angle=summary["Right Knee Angle Avg"],
    asymmetry=summary["Asymmetry Avg (%)"],
    contact_angle=summary["Contact Angle Avg"],
    max_acceleration=summary["Max Acceleration"],
    step_frequency=summary["Step Frequency"]
)

summary["Risk Level"] = risk["level"]
summary["Risk Score"] = risk["score"]

injury_tags = suspected_injury_type(
    summary["Right Knee Angle Avg"], summary["Asymmetry Avg (%)"], summary["Contact Angle Avg"],
    summary["Max Acceleration"], summary["Step Frequency"], summary["Trunk Lean Avg"], summary["Hip Flexion Avg"]
)

# ✅ 10. Display Results
summary_df = pd.DataFrame.from_dict(summary, orient='index', columns=['Value'])
output_text = summary_df.to_string()

print("\n📊 Biomechanical Summary:")
print(output_text)

print("\n📝 Detailed Notes:")
for note in risk["notes"]:
    print("-", note)

print("\n🧬 Likely Associated Injuries:")
for tag in injury_tags:
    print("-", tag)
```

# الشخص الصح في المكان الصح- The Right Person in the Right Place

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestClassifier
from typing import Dict, Tuple

# === Step 1: Load Player Data ===
def load_player_data(file_path: str = "/content/players_data_light-2024_2025.csv") -> pd.DataFrame:
    return pd.read_csv(file_path)

# === Step 2: Preprocess Data ===
def preprocess_player_data(df: pd.DataFrame) -> Tuple[np.ndarray, pd.Series, MinMaxScaler, list]:
    df = df[df['Pos'].isin(['DF', 'MF', 'FW'])].copy()

    features = ['Tkl', 'TklW', 'Int', 'Clr', 'Blocks',
                'Att', 'Cmp%', 'KP', '1/3', 'PPA', 'PrgP',
                'Touches', 'Carries', 'Gls', 'Sh', 'SoT',
                'xG', 'npxG', 'Succ']

    available_features = [f for f in features if f in df.columns]
    df = df[['Pos'] + available_features].dropna()

    X = df[available_features]
    y = df['Pos']

    scaler = MinMaxScaler()
    X_scaled = scaler.fit_transform(X)

    return X_scaled, y, scaler, available_features

# === Step 3: Train Model ===
def train_position_model(X: np.ndarray, y: pd.Series) -> RandomForestClassifier:
    model = RandomForestClassifier(n_estimators=100, random_state=42)
    model.fit(X, y)
    return model

# === Step 4: Predict Position ===
def predict_position(user_input: Dict[str, float],
                     model: RandomForestClassifier,
                     scaler: MinMaxScaler,
                     feature_order: list) -> str:
    input_values = [user_input.get(f, 0.0) for f in feature_order]
    input_array = np.array(input_values).reshape(1, -1)
    input_scaled = scaler.transform(input_array)
    return model.predict(input_scaled)[0]

# === Step 5: Get User Input ===
def get_input_from_scout(feature_list: list) -> Dict[str, float]:
    print("📋 Please enter the following player stats:")
    user_input = {}
    for feat in feature_list:
        while True:
            try:
                value = float(input(f"{feat}: "))
                user_input[feat] = value
                break
            except ValueError:
                print("⚠️ Please enter a valid number.")
    return user_input

# === Main Program ===
if __name__ == "__main__":
    df = load_player_data()
    X_scaled, y, scaler, used_features = preprocess_player_data(df)
    model = train_position_model(X_scaled, y)

    scout_input = get_input_from_scout(used_features)
    predicted_position = predict_position(scout_input, model, scaler, used_features)

    print(f"\n🎯 Based on the stats, the best-fit position is: {predicted_position}")
```

```python
85# Define the function first
def recommend_players(df: pd.DataFrame, position: str, top_n: int = 10) -> pd.DataFrame:
    """
    Recommend top N players for a given position based on relevant stats.
    """
    from sklearn.preprocessing import MinMaxScaler

    position_features = {
        'DF': ['Tkl', 'TklW', 'Int', 'Clr', 'Blocks'],
        'MF': ['Att', 'Cmp%', 'KP', '1/3', 'PPA', 'PrgP'],
        'FW': ['Gls', 'Sh', 'SoT', 'xG', 'npxG', 'Succ']
    }

    df_pos = df[df['Pos'].str.contains(position)].copy()
    features = [f for f in position_features[position] if f in df_pos.columns]
    df_pos = df_pos[['Player', 'Squad', 'Age', 'Pos'] + features].dropna()

    scaler = MinMaxScaler()
    df_pos_scaled = df_pos.copy()
    df_pos_scaled[features] = scaler.fit_transform(df_pos[features])

    df_pos_scaled['Score'] = df_pos_scaled[features].sum(axis=1)
    top_recommendations = df_pos_scaled.sort_values(by='Score', ascending=False).head(top_n)

    return top_recommendations[['Player', 'Squad', 'Age', 'Pos', 'Score']]


player_df = load_player_data()
recommend_players(df=player_df, position='MF', top_n=10)
```
